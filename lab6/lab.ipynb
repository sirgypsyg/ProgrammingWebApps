```python
import socket

host = "httpbin.org"
port = 80
request = (
    "GET /html HTTP/1.1\r\n"
    "Host: httpbin.org\r\n"
    "User-Agent: Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) "
    "AppleWebKit/537.75.14 (KHTML, like Gecko) Version/7.0.3 Safari/7046A194A\r\n"
    "Connection: close\r\n"
    "\r\n"
)

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.connect((host, port))
    s.sendall(request.encode())

    response = b""
    while True:
        data = s.recv(4096)
        if not data:
            break
        response += data

header, _, body = response.partition(b"\r\n\r\n")

with open("output.html", "wb") as f:
    f.write(body)
```

```python
import socket

def get_range(host, path, start, end):
    request = (
        f"GET {path} HTTP/1.1\r\n"
        f"Host: {host}\r\n"
        f"User-Agent: Mozilla/5.0\r\n"
        f"Accept: image/jpeg\r\n"
        f"Range: bytes={start}-{end}\r\n"
        f"Connection: close\r\n"
        "\r\n"
    )

    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((host, 80))
        s.sendall(request.encode())
        response = b""
        while True:
            data = s.recv(4096)
            if not data:
                break
            response += data

    _, _, body = response.partition(b"\r\n\r\n")
    return body

host = "progress.ski"
path = "/image.jpg"

part1 = get_range(host, path, 0, 9999)
part2 = get_range(host, path, 10000, 19999)
part3 = get_range(host, path, 20000, 29999)

full_image = part1 + part2 + part3

with open("parts.jpg", "wb") as f:
    f.write(full_image)

```

```python
import socket

def get_headers(host, path):
    request = (
        f"HEAD {path} HTTP/1.1\r\n"
        f"Host: {host}\r\n"
        f"Connection: close\r\n"
        "\r\n"
    )
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((host, 80))
        s.sendall(request.encode())
        response = b""
        while True:
            data = s.recv(4096)
            if not data:
                break
            response += data
    headers = response.decode().split("\r\n")
    header_dict = {}
    for h in headers:
        if ": " in h:
            key, value = h.split(": ", 1)
            header_dict[key.lower()] = value
    return header_dict

def get_range_if_modified(host, path, start, end, last_modified):
    request = (
        f"GET {path} HTTP/1.1\r\n"
        f"Host: {host}\r\n"
        f"User-Agent: Mozilla/5.0\r\n"
        f"Accept: image/jpeg\r\n"
        f"Range: bytes={start}-{end}\r\n"
        f"If-Modified-Since: {last_modified}\r\n"
        f"Connection: close\r\n"
        "\r\n"
    )

    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((host, 80))
        s.sendall(request.encode())
        response = b""
        while True:
            data = s.recv(4096)
            if not data:
                break
            response += data

    if b"304 Not Modified" in response:
        print("Obrazek nie został zmieniony.")
        return None

    _, _, body = response.partition(b"\r\n\r\n")
    return body

host = "progress.ski"
path = "/image.jpg"

# Pobierz ostatnią datę modyfikacji
headers = get_headers(host, path)
last_modified = headers.get("last-modified")

# Jeśli nie ma nagłówka, pobierz bez warunku
if not last_modified:
    last_modified = ""

part1 = get_range_if_modified(host, path, 0, 9999, last_modified)
part2 = get_range_if_modified(host, path, 10000, 19999, last_modified)
part3 = get_range_if_modified(host, path, 20000, 29999, last_modified)

if part1 and part2 and part3:
    full_image = part1 + part2 + part3
    with open("parts.jpg", "wb") as f:
        f.write(full_image)
else:
    print("Użyto cache — plik nie został pobrany ponownie.")
```

```python
import socket

def handle_request(request):
    try:
        lines = request.split("\r\n")
        request_line = lines[0]
        method, path, _ = request_line.split()

        if method == "GET":
            if path == "/":
                return "HTTP/1.1 200 OK\r\nContent-Type: text/html\r\n\r\n<h1>Witaj na serwerze HTTP!</h1>"
            elif path == "/test":
                return "HTTP/1.1 200 OK\r\nContent-Type: text/html\r\n\r\n<h1>Testowy zasób!</h1>"
            else:
                return "HTTP/1.1 404 Not Found\r\nContent-Type: text/html\r\n\r\n<h1>404 - Zasób nie znaleziony</h1>"
        else:
            return "HTTP/1.1 405 Method Not Allowed\r\nContent-Type: text/html\r\n\r\n<h1>405 - Metoda niedozwolona</h1>"
    
    except Exception as e:
        return f"HTTP/1.1 500 Internal Server Error\r\nContent-Type: text/html\r\n\r\n<h1>500 - Błąd serwera: {str(e)}</h1>"

def run_server(host, port):
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind((host, port))
    server_socket.listen(1)
    print(f"Serwer działa na {host}:{port}...")

    while True:
        client_socket, client_address = server_socket.accept()
        
        try:
            request = client_socket.recv(1024).decode("utf-8")
            response = handle_request(request)
            client_socket.sendall(response.encode("utf-8"))
        except Exception as e:
            print(f"Błąd przetwarzania żądania: {e}")
        finally:
            client_socket.close()

if __name__ == "__main__":
    run_server("127.0.0.1", 8080)
```
